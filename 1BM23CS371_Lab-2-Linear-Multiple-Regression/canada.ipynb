```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('canada_per_capita_income.csv')

# Check data
print(df.head())

# Define X (year) and y (income)
X = df[['year']]        # independent variable
y = df['per capita income (US$)']   # dependent variable

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict income for year 2020
prediction_2020 = model.predict(pd.DataFrame({'year':[2020]}))

print("Predicted per capita income in 2020:", prediction_2020[0])
```
