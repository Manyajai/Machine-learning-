```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('salary.csv')

# Check data
print(df.head())

# Define X and y
X = df[['YearsExperience']]
y = df['Salary']

# Create and train model
model = LinearRegression()
model.fit(X, y)

# Predict salary for 12 years experience
prediction_salary = model.predict(pd.DataFrame({'YearsExperience':[12]}))

print("Predicted salary for 12 years experience:", prediction_salary[0])
```
