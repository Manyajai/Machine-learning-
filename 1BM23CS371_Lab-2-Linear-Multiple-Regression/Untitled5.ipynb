```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('1000_Companies.csv')

print(df.head())

# Convert categorical variable (State) into dummy variables
df = pd.get_dummies(df, drop_first=True)

# Define X and y
X = df.drop('Profit', axis=1)
y = df['Profit']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict for:
# 91694.48 R&D, 515841.3 Admin, 11931.24 Marketing, Florida

prediction = model.predict([[91694.48, 515841.3, 11931.24, 1, 0]])

print("Predicted Profit:", prediction[0])
```
