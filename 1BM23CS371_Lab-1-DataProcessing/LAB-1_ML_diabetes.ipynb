```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
```

```python
diabetes = pd.read_csv("Dataset of Diabetes .csv")
```

```python
diabetes.isnull().sum()
```

```python
le = LabelEncoder()
diabetes['Gender'] = le.fit_transform(diabetes['Gender'])
diabetes['CLASS'] = le.fit_transform(diabetes['CLASS'])
```

```python
Q1 = diabetes.quantile(0.25)
Q3 = diabetes.quantile(0.75)
IQR = Q3 - Q1

diabetes = diabetes[~((diabetes < (Q1 - 1.5 * IQR)) |
                      (diabetes > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
mm = MinMaxScaler()
diabetes_minmax = mm.fit_transform(diabetes.drop(['ID','No_Pation'], axis=1))
```

```python
ss = StandardScaler()
diabetes_standard = ss.fit_transform(diabetes.drop(['ID','No_Pation'], axis=1))
```
